## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Code to make dataset analysis ready.


# Filling age_certification column with mode
if "age_certification" in df.columns:
    df["age_certification"] = df["age_certification"].fillna(df["age_certification"].mode()[0])

# Filling cast-related missing values
for col in ["name", "character"]:
    if col in df.columns:
        df[col] = df[col].fillna("Unknown")

# Filling season column missing values with mean
if "seasons" in df.columns:
    df["seasons"] = df["seasons"].fillna(df["seasons"].mean())

# Filling rating & score columns with median
num_cols = ["imdb_score", "imdb_votes", "tmdb_score", "tmdb_popularity"]
for col in num_cols:
    if col in df.columns:
        df[col] = df[col].fillna(df[col].median())

# drop null values in imdb_id
df = df.dropna(subset=['imdb_id'])

# drop null values in descripton column
df = df.dropna(subset=['description'])
````
